# Nintendo Analyses

##### In this analyse We'll consider only the games published by Nintendo for Nintendo consoles

```python
# Here We're going to create a variable and insert a query to see the number of sales in each platform by Nintendo.
# The number of sales is based in  Global sales,so we have the entire number of copies around the world.
# It's possible to see that Nintendo Wii has more than the double of the copies sales made for Nes(Nintendo Entertainment System)
# Another pattern we have here is the portable consoles from nintendo has a considerable number of sales if we compare to the desk consoles.

nintendo = games.query('Publisher == "Nintendo"').groupby("Platform")["Global_Sales"].sum()
```

```python
nintendo.sort_values(ascending = False)
```

```python
# Here We want to know the global sales by genre, when the publisher is Nintendo.
# It's using command query and grouping by genre and sum up the global sales, ordering decrescent
# It created a variable to make easier if we need to select those informations again.

nintendo_genre = games.query('Publisher == "Nintendo"').groupby("Genre")["Global_Sales"].sum().sort_values(ascending = False)
```

```python
# In the result is possible to assess that when Nintendo publishes a game by itself, they'd rather to work on platform games, such as:
# Mario, Donkey Kong  Country and Metroid.
# The second position is RPG(Role Playing Games) and had a considerable number of sales too. They provided games such:
# Zelda, Mario RPG, and the most famous in their portable consoles called Pokemon.

nintendo_genre
```
